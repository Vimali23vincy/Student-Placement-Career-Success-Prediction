```python
df.drop(['age','gender','primary_language','attendance_percentage'], axis=1, inplace=True)
df.head()
```

```python
df["placement_status"] = df["placement_status"].map({"Placed": 1,"Not Placed": 0})
df.head()
```

# MODEL 1

**TARGET 1**

Placement_status

```python
x = df.drop(['placement_status','company_type'], axis=1)
y = df['placement_status']
```

```python
x=pd.get_dummies(x,dtype='int')
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=42)
```

**LOGISTIC REGRESSION**




```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy_score(y_test,y_pred)
```

```python
confus_mat=confusion_matrix(y_test,y_pred)
confus_mat
```

```python
classification_rep=classification_report(y_test,y_pred)
print(classification_rep)
```

**RANDOMFOREST CLASSIFIER**


```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=100,random_state=42)
model.fit(x_train, y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
print("Test Accuracy:", accuracy_score(y_test, y_pred))
```

```python
classification_rep=print(classification_report(y_test, y_pred))
print(classification_rep)
```

```python
from sklearn.metrics import confusion_matrix
confus_mat=confusion_matrix(y_test, y_pred)
confus_mat
```

```python
import joblib
joblib.dump(model, "placement_status_model.pkl")
```

# **MODEL 2**

COMPANY_TYPE

```python
placed_df = df[df["placement_status"] == 1]
placed_df.head()
```

```python
x = placed_df.drop(["placement_status", "company_type"],axis=1)
y = placed_df["company_type"]
```

```python
x=pd.get_dummies(x,dtype='int')
x.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

```python
confus_mat=confusion_matrix(y_test, y_pred)
confus_mat
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=42)
rf.fit(x_train, y_train)
```

```python
y_pred = rf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

```python
confus_mat=confusion_matrix(y_test, y_pred)
confus_mat
```

```python
joblib.dump(rf, "company_type_model.pkl")
```
